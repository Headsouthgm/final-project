```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
chip = pd.read_csv("crime.csv")
crime1 = chip['INCIDENTHIERARCHYDESC'] == "AGG ASSAULT/OTHER/INJURY"
crime2 = chip['INCIDENTHIERARCHYDESC'] == "AGG ASSAULT/KNIFE/INJURY"
crime3 = chip['INCIDENTHIERARCHYDESC'] == "AGG ASSAULT/STRONGARM/INJURY"
crime4 = chip['INCIDENTHIERARCHYDESC'] == "AGG ASSAULT/GUN/NO INJURY"
crime5 = chip['INCIDENTHIERARCHYDESC'] == "AGG ASSAULT/GUN/INJURY"
crime6 = chip['INCIDENTHIERARCHYDESC'] == "AGG ASSAULT/KNIFE/NO INJURY"

neigh1 = chip['INCIDENTNEIGHBORHOOD'] == "Mt. Oliver Boro" 
neigh2 = chip['INCIDENTNEIGHBORHOOD'] == "Outside County"
neigh3 = chip['INCIDENTNEIGHBORHOOD'] == "Mt. Oliver Neighborhood"
neigh4 = chip['INCIDENTNEIGHBORHOOD'] == "Chartiers City"

neigh1_crime1 = chip[crime1 & neigh1].value_counts()
neigh1_crime2 = chip[crime2 & neigh1].value_counts()
neigh1_crime3 = chip[crime3 & neigh1].value_counts()
neigh1_crime4 = chip[crime4 & neigh1].value_counts()
neigh1_crime5 = chip[crime5 & neigh1].value_counts()
neigh1_crime6 = chip[crime6 & neigh1].value_counts()
num11 = len(neigh1_crime1)
num12 = len(neigh1_crime2)
num13 = len(neigh1_crime3)
num14 = len(neigh1_crime4)
num15 = len(neigh1_crime5)
num16 = len(neigh1_crime6)

neigh1_sumcrime = num11 + num12+num13+num14+num15+num16
print("The number of serious crime in Mt. Oliver Boro: " + str(neigh1_sumcrime))
neigh2_crime1 = chip[crime1 & neigh2].value_counts()
neigh2_crime2 = chip[crime2 & neigh2].value_counts()
neigh2_crime3 = chip[crime3 & neigh2].value_counts()
neigh2_crime4 = chip[crime4 & neigh2].value_counts()
neigh2_crime5 = chip[crime5 & neigh2].value_counts()
neigh2_crime6 = chip[crime6 & neigh2].value_counts()
num21 = len(neigh2_crime1)
num22 = len(neigh2_crime2)
num23 = len(neigh2_crime3)
num24 = len(neigh2_crime4)
num25 = len(neigh2_crime5)
num26 = len(neigh2_crime6)
neigh2_sumcrime = num21 + num22+num23+num24+num25+num26
print("The number of serious crime in Outside County: " + str(neigh2_sumcrime))

neigh3_crime1 = chip[crime1 & neigh3].value_counts()
neigh3_crime2 = chip[crime2 & neigh3].value_counts()
neigh3_crime3 = chip[crime3 & neigh3].value_counts()
neigh3_crime4 = chip[crime4 & neigh3].value_counts()
neigh3_crime5 = chip[crime5 & neigh3].value_counts()
neigh3_crime6 = chip[crime6 & neigh3].value_counts()
num31 = len(neigh3_crime1)
num32 = len(neigh3_crime2)
num33 = len(neigh3_crime3)
num34 = len(neigh3_crime4)
num35 = len(neigh3_crime5)
num36 = len(neigh3_crime6)
neigh3_sumcrime = num31 + num32 + num33+ num34 + num35 + num36
print("The number of serious crime in Mt. Oliver Neighborhood: " + str(neigh3_sumcrime))

neigh4_crime1 = chip[crime1 & neigh4].value_counts()
neigh4_crime2 = chip[crime2 & neigh4].value_counts()
neigh4_crime3 = chip[crime3 & neigh4].value_counts()
neigh4_crime4 = chip[crime4 & neigh4].value_counts()
neigh4_crime5 = chip[crime5 & neigh4].value_counts()
neigh4_crime6 = chip[crime6 & neigh4].value_counts()
num41 = len(neigh3_crime1)
num42 = len(neigh3_crime2)
num43 = len(neigh3_crime3)
num44 = len(neigh3_crime4)
num45 = len(neigh3_crime5)
num46 = len(neigh3_crime6)
neigh4_sumcrime = num41 + num42 + num43+ num44 + num45 + num46
print("The number of serious crime in Chartiers City: "+ str(neigh4_sumcrime))

```

```python

import pandas as pd
df = pd.read_csv('value.csv')

# Filter the DataFrame to find the price of neighborhood 'A'
neighborhood_Mt_Oliver_df = df[df['Neighborhood'] == 'Mt. Oliver']

# Check if the filtered DataFrame is not empty
if not neighborhood_Mt_Oliver_df.empty:
    # Extract the price value from the filtered DataFrame
    neighborhood_Mt_Oliver_price = neighborhood_Mt_Oliver_df['Estimate; Total:'].iloc[0]
    print("Price of neighborhood Mt Oliver:", neighborhood_Mt_Oliver_price)
else:
    print("Neighborhood Mt Oliver not found in the data.")
```
